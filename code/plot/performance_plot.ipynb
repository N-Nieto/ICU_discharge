# ICU ML performance

```python
import scipy.stats as stats
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
from pathlib import Path


output_dir = Path().resolve().parents[1] / "output"
results_df = pd.read_csv(
    output_dir / "LG" / "icu_discharge_VP_model_lg.csv", index_col=0
)

plt.figure(figsize=(10, 7))
results_df["Model"].replace(
    {
        "ICU model (LG+imputed)": "ICU discharge",
        "ICU model (LG+imputed) Train": "ICU discharge Train",
    },
    inplace=True,
)

sbn.lineplot(data=results_df, y="Balanced ACC", x="Thresholds", hue="Model")
plt.grid(color="black", alpha=0.5)
plt.title("ICU discharge - LG model performance")
plt.xlim([0, 1])
plt.show()
# %%
```

```python
results_2 = results_df[results_df.Thresholds == 0.5]
sbn.boxplot(data=results_2, y="AUC", x="Model")
plt.grid(axis="y", color="black", alpha=0.5)
plt.title("AUC performance of LG model")
```

```python
results_true_label = results_df[results_df.Thresholds == 0.5]
metrics = ["Balanced ACC", "AUC", "Sensitivity", "Specificity", "F1"]
model = "ICU discharge"
results_true_label = results_true_label[results_true_label["Thresholds"] == 0.5]
for metric in metrics:
    results_mean = round(
        results_true_label[results_true_label["Model"] == model][metric].mean(), 3
    )
    results_std = round(
        results_true_label[results_true_label["Model"] == model][metric].std(), 3
    )
    confidence_interval = stats.t.interval(
        0.95,
        len(results_true_label[results_true_label["Model"] == model][metric]) - 1,
        loc=results_mean,
        scale=results_std,
    )

    print(
        metric
        + ": "
        + str(results_mean)
        + " ["
        + str(round(confidence_interval[0], 3))
        + "/"
        + str(round(confidence_interval[1], 3))
        + "]"
    )

# %%
```
